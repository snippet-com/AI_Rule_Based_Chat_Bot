```python
def get_response(user_input):
    """
    Processes user input and returns appropriate response based on predefined rules.
    """
    user_input = user_input.lower().strip()

    # Greeting handling
    if user_input in ["hi", "hello", "hey"]:
        greetings = [
            "Hello! How can I help you?",
            "Hi there!",
            "Hey! Nice to meet you."
        ]
        return greetings[hash(user_input) % len(greetings)]

    # Exit commands
    if user_input in ["bye", "exit", "quit"]:
        return "Goodbye!"

    # Question-response pairs
    responses = {
        "how are you": "I am fine.",
        "what is your name": "I am a rule-based chatbot.",
        "who made you": "I was created using Python.",
        "help": "You can ask simple questions."
    }

    if user_input in responses:
        return responses[user_input]

    # Unknown input handling
    return "Sorry, I do not understand that."


```

```python
def main():
    """
    Main function to run the chatbot.
    """
    print("=" * 50)
    print("Hello! I am your chatbot.")
    print("Type 'bye', 'exit', or 'quit' to end the chat.")
    print("=" * 50)

    while True:
        user_input = input("\nYou: ").strip()

        if not user_input:
            continue

        response = get_response(user_input)
        print(f"Bot: {response}")

        # Check if user wants to exit
        if user_input.lower() in ["bye", "exit", "quit"]:
            break

    print("\nChat ended.")


```

```python
#main function
if __name__ == "__main__":
    main()
```
